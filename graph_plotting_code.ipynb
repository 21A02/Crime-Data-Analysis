```python
from plotly.offline import plot,iplot
import plotly.graph_objs as go
import numpy as np
```

```python
df.head(50)
```

```python
#df1=df.head(500)
df1=df
```

```python
def states_vs_crimerates_plot(crime):
    trace = go.Bar( 
        x = df1['STATE/UT'],
        y = df1[crime],
        name= 'DATA')
    data = [trace]

    fig = go.Figure(data=data)
    iplot(fig, filename='states')
    
def year_vs_crimerates_plot(crime):
    trace = go.Bar( 
        x = df1['YEAR'],
        y = df1[crime],
        name= 'DATA')
    data = [trace]

    fig = go.Figure(data=data)
    iplot(fig, filename='testing')
```

```python
states_vs_crimerates_plot('MURDER')
```

```python
year_vs_crimerates_plot('RAPE')
```

```python
def districts_vs_crimerates(state,crime):
    df2=df.loc[df['STATE/UT'] == state]
    df1=df2.drop(['STATE/UT'], axis=1)
    df1 = df1.set_index("DISTRICT")
    df1 = df1.drop("TOTAL", axis=0)
    df1.reset_index(inplace = True) 
    trace = go.Bar( 
        x = df1['DISTRICT'],
        y = df1[crime],
        name= 'DATA')
    data = [trace]

    fig = go.Figure(data=data)
    iplot(fig, filename='districts')
    
```

```python
districts_vs_crimerates('ANDHRA PRADESH','RAPE')
```
